### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
# Calcular el total de pedidos y repeticiones por producto
total_pedidos = order_products.groupby('product_id')['order_id'].nunique()
repeticiones_por_producto = order_products.groupby('product_id')['reordered'].sum()
```

```python
# Calcular la tasa de repetición por producto
tasa_repeticion_producto = repeticiones_por_producto / total_pedidos

# Mostrar las primeras 20 tasas de repetición
tasa_repeticion_producto.head(20)
```

Esto refleja la lealtad de los clientes hacia ciertos productos, que se repiten constantemente en sus pedidos.*
